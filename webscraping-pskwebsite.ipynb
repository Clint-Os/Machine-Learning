```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://psk.or.ke/membership/members-list'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find_all('table')
```

```python
print(table)
```

```python
titles = soup.find_all('th')
```

```python
titles
```

```python
psk_titles = [title.text.strip() for title in titles]

print(psk_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = psk_titles)

df
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    print(ind_row_data)

```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = ind_row_data
```
